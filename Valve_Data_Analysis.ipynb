```python
#Chi-Square test to determine consecutive number of days
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

# Load your data
data = pd.read_excel('Valve_Data_Analysis.xlsx')

# Function to create the consecutive days feature
def create_consecutive_days_feature(df, days_threshold):
    df['Increasing_Error'] = df['Error'].diff() > 0
    df['Consecutive_Days'] = df['Increasing_Error'].rolling(window=days_threshold).sum()
    df['Failure'] = (df['Consecutive_Days'] >= days_threshold).astype(int)
    return df

# Define thresholds to test
thresholds = [3, 4, 5]
results = []

for threshold in thresholds:
    df = create_consecutive_days_feature(data.copy(), threshold)
    
    # Create contingency table
    contingency_table = pd.crosstab(df['Failure'], df['Consecutive_Days'] >= threshold)
    
    # Perform chi-square test
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    
    # Store the results
    results.append({
        'Threshold': threshold,
        'Chi-Square': chi2,
        'P-value': p
    })

# Create a DataFrame to display the results
results_df = pd.DataFrame(results)

# Print the results
print(results_df)

# Choose the optimal threshold based on the p-value
optimal_threshold = results_df.loc[results_df['P-value'].idxmin(), 'Threshold']
print(f"The optimal number for consecutive days is {optimal_threshold} days.")
```

```python
import warnings
# Suppress all warnings
warnings.filterwarnings('ignore')
# Function to compute daily rolling mean

def compute_daily_rolling_mean(df):
    df['Error_Rolling_Mean'] = abs(df['Error'].rolling(window=144).mean())
    df['Date_Only'] = df['Data'].dt.date
    daily_rolling_mean = df.groupby('Date_Only')['Error_Rolling_Mean'].mean().reset_index()
    return daily_rolling_mean

# Algorithm to predict the valve failure
def detect_failure(df, threshold):
    df['Failure'] = (df['Error_Rolling_Mean'] > threshold).astype(int)

    consecutive_days = 3
    consecutive_failures = 0
    consecutive_failures_list = []

    for index, row in df.iterrows():
        if row['Failure'] == 1:
            consecutive_failures += 1
        else:
            consecutive_failures = 0
        consecutive_failures_list.append(consecutive_failures)

    df['Consecutive_Failures'] = consecutive_failures_list
    df['Predicted_Failure'] = (df['Consecutive_Failures'] >= consecutive_days).astype(int)
    return df

# Process each Excel file
for file in excel_files:
    # Load the valve data
    valve_data = pd.read_excel(file)

    # Convert 'Date' column to datetime format
    valve_data['Data'] = pd.to_datetime(valve_data['Data'])

    # Compute daily rolling mean
    daily_rolling_mean = compute_daily_rolling_mean(valve_data)

    # Apply the failure detection function to your daily rolling mean DataFrame
    daily_rolling_mean = detect_failure(daily_rolling_mean, abs(threshold))

    # Filter and get only the dates where 'Predicted_Failure' is 1
    predicted_failure_dates = daily_rolling_mean[daily_rolling_mean['Predicted_Failure'] == 1]['Date_Only']

    # Ensure the dates are in datetime format before formatting
    predicted_failure_dates = pd.to_datetime(predicted_failure_dates)

    # Convert date format to DD-MM-YYYY
    predicted_failure_dates = predicted_failure_dates.dt.strftime('%d-%m-%Y')

    # Print the results
    base_filename = os.path.splitext(os.path.basename(file))[0]
    if not predicted_failure_dates.empty:
        print(f"The valve {base_filename} failed on the following days")
        for date in predicted_failure_dates:
            print(date)
    else:
        print(f"The valve {base_filename} didn't fail.")
```
